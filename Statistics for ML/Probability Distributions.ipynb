# üìä Probability Distributions  

Probability distributions describe how values of a random variable are distributed. They are classified into **discrete** and **continuous** distributions.

## 1Ô∏è‚É£ Normal Distribution (Gaussian)  

A **Normal Distribution** is a symmetric, bell-shaped curve where most values cluster around the mean.

### üî¢ Probability Density Function (PDF):
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

### üìå Where:
- \( \mu \) = Mean  
- \( \sigma \) = Standard Deviation  

### üìù Key Properties:
- Symmetric about \( \mu \)  
- Mean = Median = Mode  
- Used in real-world applications like IQ scores, stock returns  

---

## 2Ô∏è‚É£ Binomial Distribution  

A **Binomial Distribution** models the number of successes in \( n \) independent Bernoulli trials.

### üî¢ Probability Mass Function (PMF):
$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n-k}
$$

### üìå Where:
- \( n \) = Total trials  
- \( k \) = Number of successes  
- \( p \) = Probability of success in one trial  

### üìù Key Properties:
- Used for binary outcomes (success/failure)  
- Example: Tossing a coin \( n \) times and counting heads  

---

## 3Ô∏è‚É£ Poisson Distribution  

A **Poisson Distribution** models the number of occurrences in a fixed interval of time or space.

### üî¢ Probability Mass Function (PMF):
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

### üìå Where:
- \( \lambda \) = Average rate of occurrence  
- \( k \) = Number of occurrences  

### üìù Key Properties:
- Used for rare events (e.g., number of customer arrivals per hour)  
- The mean and variance are both \( \lambda \)  

---

## 4Ô∏è‚É£ Bernoulli Distribution  

A **Bernoulli Distribution** models a single experiment with only two outcomes: success (1) or failure (0).

### üî¢ Probability Mass Function (PMF):
$$
P(X = x) = p^x (1 - p)^{1-x}, \quad x \in \{0,1\}
$$

### üìå Where:
- \( p \) = Probability of success  
- \( 1 - p \) = Probability of failure  

### üìù Key Properties:
- Special case of the binomial distribution with \( n = 1 \)  
- Example: Tossing a coin once  

---

## 5Ô∏è‚É£ Log-Normal Distribution  

A **Log-Normal Distribution** describes a variable whose logarithm follows a normal distribution.

### üî¢ Probability Density Function (PDF):
$$
f(x) = \frac{1}{x\sigma \sqrt{2\pi}} e^{-\frac{(\ln x - \mu)^2}{2\sigma^2}}, \quad x > 0
$$

### üìå Where:
- \( \mu \) = Mean of the log-transformed variable  
- \( \sigma \) = Standard deviation of the log-transformed variable  

### üìù Key Properties:
- Skewed distribution (not symmetric)  
- Used for modeling stock prices, income distribution  

---

## üî• **Understanding PDF vs PMF**
- **Probability Density Function (PDF)**: Used for continuous distributions (e.g., Normal, Log-Normal).  
- **Probability Mass Function (PMF)**: Used for discrete distributions (e.g., Binomial, Poisson, Bernoulli).
